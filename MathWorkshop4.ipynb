```python
import matplotlib.pyplot as plt
Bear = [8, 54, 93, 116]
Dolphin = [150, 77, 32, 11]
Whale = [80, 54, 100, 76]
year = 2017
groups=['Bears','Dolphin','Whale']
for i in range (len(Bear)):
  values = [Bear[i], Dolphin[i], Whale[i]]
  plt.pie(values, labels=groups)
  plt.title(f"{2017 + i}")
  plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
data = [23, 45, 56, 78, 213]
plt.bar(x, data,width=0.5,color='maroon')
plt.title("A bar plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show
plt.grid(color='#95a5a6', linestyle='--',linewidth=2, axis='y',
alpha=0.7)
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
data = [123, 230, 78, 193, 12]
names = ["Red", "Silver", "Blue", "White", "Green"]
colors = ["red", "silver", "blue", "white", "green"]
plt.bar(x, data, width=0.5, color=colors)

plt.title("Colors Seen")
plt.xlabel("Colors")
plt.ylabel("Numbers")
plt.xticks(x, names)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.show()
```
