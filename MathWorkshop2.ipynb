```python
def unique_list(lst):
    result = []
    for i in lst:
        if i not in result:
            result.append(i)
    return result

print(unique_list([1,2,3,3,3,3,4,5]))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for i in range(3, 1000, 2):
    if is_prime(i) and is_prime(i+2):
        print(i, i+2)
```

```python
def sumOfDivisors(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s

n = 7
while True:
    if sumOfDivisors(n) == n:
        print(n)
        break
    n += 1
```

```python
marks = [20, 30, 35, 55, 70, 80, 90]

total = 0
for i in marks:
    total += i
mean = total / len(marks)

n = len(marks)
if n % 2 != 0:
    median = marks[n//2]
else:
    median = (marks[n//2 - 1] + marks[n//2]) / 2

print(mean)
print(median)
```

```python
def dec_to_bin(n):
    b = ""
    while n > 0:
        b = str(n % 2) + b
        n = n // 2
    return b

print(dec_to_bin(10))
```

```python
def prodDigits(n):
    p = 1
    while n > 0:
        p *= n % 10
        n = n // 10
    return p

print(prodDigits(123))
```
