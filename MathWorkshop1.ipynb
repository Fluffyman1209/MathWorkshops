```python
#13
def square(n):
  return n**2
def sum_squares(x):
  sum = 0
  for i in range(x+1):
    sum = sum + square(i)
  return sum
print(sum_squares(3))
#
```

```python
#exercise-2:
def myPow(x,n):
  i = 1
  nv = 1
  p = n
  n = abs(n)
  while i<=n:
    nv=nv*x
    i+=1
  if(p<0):
    return 1/nv
  else:
    return nv
print(myPow(2,-3))


```
